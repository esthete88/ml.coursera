## Генерирую выборку обьема 1000, строю гистограмму и график функции плотности распределения

```python
expon_rv = sts.expon()
mean = 1
std = 1
#mean, var = expon_rv.stats(moments='mv')
print(mean,var)
r = expon.rvs(size=1000)
plt.hist(r, bins=20, histtype='stepfilled', color='pink', normed=True)
x = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 1000)
pdf = expon.pdf(x)
plt.plot(x, pdf, lw=3, color='lightgreen', label = 'Probability density function (exp)')
plt.legend(loc='upper right')
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

## Генерирую 1000 выборок обьема 5, строю гистограмму и график функции плотности соответсвующего нормального распределения

```python
means1 = []
i = 0
for i in range(0,1000):
    r1 = expon.rvs(size=5)
    m1 = np.mean(r1)
    means1.append(m1)
    i += 1
plt.hist(means1, bins=20, histtype='stepfilled', color='pink', normed=True)
#norm_rv1 = sts.norm(mean, var)
mean1=np.mean(means1)
std1=np.std(means1)
print(mean1, std1)
norm_rv1 = sts.norm(mean1, std1)
pdf1 = norm_rv1.pdf(x)
plt.plot(x, pdf1, lw=4, color='lightgreen', label = 'Probability density function (norm)')
plt.legend(loc='upper right')
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```
