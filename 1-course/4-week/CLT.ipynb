# Центральная предельная теорема своими руками

Будем использовать двойное распределение Вейбулла

```python
import numpy as np
import scipy.stats as st
import seaborn
%pylab inline
```

```python
distr = st.dweibull(2.07) # инициализация распределения
rng = np.linspace(-3, 3, 1000)
frng = distr.pdf(rng)
```

Математическое ожидание и дисперсия:

```python
mu = distr.stats()[0]
sigma = np.sqrt(distr.stats()[1])
mu, sigma
```

Сгенерируем выборку из 1000

```python
a = distr.rvs(1000)
```

```python
hist(a, normed=True, bins=30) #bins - кол-во столбцов в гистограмме
plot(rng, frng, 'y')
xlabel('$x$')
ylabel('$f(x)$')
show()
```

```python
Sample = np.zeros(1000)
```

Построим гистограммы и теоретические функции плотности вероятности для разных размеров выборки (параметр size)

```python
size = 2
rng = np.linspace(-3, 3, 1000)
for i in range(0, 1000):
    Sample[i] = distr.rvs(size).mean()
norm_rv = st.norm(loc=mu, scale=sigma/np.sqrt(size))
plot(rng, norm_rv.pdf(rng), label='$\mu=%s, \sigma^2=%s$' % (mu, sigma**2/size))
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
hist(Sample, normed=True, bins=50)
xlabel('$x$')
ylabel('$f(x)$')
title('Размер выборки = 2')
show()
```
