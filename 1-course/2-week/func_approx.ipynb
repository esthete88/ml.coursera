```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy as sp

f = lambda x: np.sin(x/5.0)*np.exp(x/10.0) + 5*np.exp(-x/2.0)

arr = np.arange(0, 15, 0.01)
plt.plot(arr, f(arr))
```

```python
sample = [1, 15]
A = np.array([[x, 1] for x in sample])
b = np.array([f(x) for x in sample])
s = np.linalg.solve(A, b)
plt.plot(arr, f(arr))
plt.plot(arr, np.polyval(s, arr))
```

```python
sample = [1, 8, 15]
A = np.array([[x ** 2, x, 1] for x in sample])
b = np.array([f(x) for x in sample])
s = np.linalg.solve(A, b)
plt.plot(arr, f(arr))
plt.plot(arr, np.polyval(s, arr))
```

```python
sample = [1, 4, 10, 15]
A = np.array([[x ** 3, x ** 2, x, 1] for x in sample])
b = np.array([f(x) for x in sample])
s = np.linalg.solve(A, b)
plt.plot(arr, f(arr))
plt.plot(arr, np.polyval(s, arr))
```
